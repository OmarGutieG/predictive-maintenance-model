```python
# Dropping missing values, reseting indexes (important for next steps) and printing new shape
df = df.dropna(axis=0)
df.reset_index(drop=True, inplace=True)
df.shape
```

## Scaling data

```python
# The only feature we need to scale is the ´days_since_last_performed' since everything else is normalized with 0s and 1s
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))

# Fit and transform the selected column
column_values = df['days_since_last_performed'].values.reshape(-1, 1)
df['days_since_last_performed'] = scaler.fit_transform(column_values)

# Display the normalized DataFrame
df['days_since_last_performed']
```

## Dealing with categorical data

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)

columns_order = df.columns.tolist()

categorical_cols = [
    'asset',
    'asset_type',
    'asset_make',
    'asset_model',
    'industry',
    'store_performer',
    'inspection_type',
    'inspector',
    'asset_owner',
    'last_performed_inspector',
    'original_inspection',
    'original_checklist',
    'last_performed_inspection_result'
]

for i, col in enumerate(categorical_cols):
    # Get the index from the column order
    index = columns_order.index(col)
    # Encode data
    encoded_data = encoder.fit_transform(df[[col]])
    # Get columns of the enconded DataFrame
    feature_names = encoder.get_feature_names_out([col])
    # Convert the encoded data to a DataFrame with column names
    encoded_df = pd.DataFrame(encoded_data, columns=feature_names)
    # Remove categorical column from original DataFrame
    df.drop([col], axis=1, inplace=True)
    # Add the new encoded columns
    df = pd.concat([df, encoded_df], axis=1)
    # Remove categorical column from the columns order
    del columns_order[index]
    # Add all the new encoded columns to the columns order at the right index
    for j, enco_col in enumerate(encoded_df.columns):
        columns_order.insert(index, enco_col)
        index += 1

# Apply correct order to DataFrame
df = df[columns_order]
print(len(columns_order))
```

## Dimentionality reduction

```python
import keras
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from sklearn.model_selection import train_test_split

X = df.iloc[:,:-3].values

# Split the data into training and testing sets
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)

# Define the architecture of the encoder
input_size = X_train.shape[1]

input_data = Input(shape=(input_size,))
encoding_dim = 1024

# Encoder layers
encoded = Dense(8192, activation='tanh')(input_data)
encoded = Dense(4096, activation='tanh')(encoded)
encoded = Dense(encoding_dim, activation='tanh')(encoded)

# Decoder layers
decoded = Dense(4096, activation='tanh')(encoded)
decoded = Dense(8192, activation='tanh')(decoded)
decoded = Dense(input_size, activation='sigmoid')(decoded)

autoencoder = Model(inputs=input_data, outputs=decoded)

# Compile the autoencoder model
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
autoencoder.summary()
```

```python
# Train the autoencoder on the categorical data
autoencoder.fit(X_train, X_train, epochs=5, batch_size=32, shuffle=True, validation_data=(X_test, X_test))
```

```python
# Evaluate the model on the test set
evaluation_results = autoencoder.evaluate(X_test, X_test)

# Display evaluation results
print(f"Loss: {evaluation_results}")
```

```python
# Create the model with only the encoding layers
encoder_model = Model(inputs=input_data, outputs=encoded)

# Encode all features
reduced_features_train = encoder_model.predict(X_train)
reduced_features_test = encoder_model.predict(X_test)
```

```python
# Create DataFrames for reduced features
reduced_df_train = pd.DataFrame(reduced_features_train, columns=[f"encoded_{i+1}" for i in range(encoding_dim)])
reduced_df_test = pd.DataFrame(reduced_features_test, columns=[f"encoded_{i+1}" for i in range(encoding_dim)])

# Concatenate reduced feature DataFrames
reduced_combined = pd.concat([reduced_df_train, reduced_df_test], axis=0, ignore_index=True)
reduced_combined
```

```python
# Choose the two columns to concatenate from the original DataFrame
columns_to_concat = ["days_since_last_performed", "inspection_result"]

# Concatenate the selected columns with the reduced_combined DataFrame
reduced_df = pd.concat([reduced_combined, df[columns_to_concat]], axis=1)
reduced_df
```
